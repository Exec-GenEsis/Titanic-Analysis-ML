## Abstract

In this project, we investigated whether passenger characteristics on the Titanic affected survival and whether these factors could be used to predict survivability. We focused on variables such as gender, age, and passenger class, especially in the context of the “Women and Children First” policy. We also wanted to determine if survival outcomes were mostly explained by these characteristics or if there was still a large element of randomness in their survivability.

To do this, we used and cleaned a dataset 891 passengers, removing missing values and preparing the data for analysis. We performed EDA and statistical tests, including chi-squared tests and t-tests, to examine relationships between variables and survival. We then built several machine learning models, including logistic regression and a support vector machine (SVM), using features like class, age, fare, and family size, and evaluated their performance using accuracy and ROC-AUC.

We found that gender and passenger class were the strongest predictors of survival. Female passengers and higher classes were significantly more likely to survive, supporting the impact of social norms and access to resources. Age had a smaller effect, with younger passengers slightly more likely to survive. Our models achieved moderate accuracy (around 78–83%), showing that survival was not random but also not perfectly predictable, suggesting that both structured factors and chance played a role.

## Data Cleaning

Before exploring and training using the dataset, we must first clean up missing values in the columns as well as remove duplicate columns.

Our reasons for this are twofold
1. Scikit's StandardScaler and encoding methods have no easy way of processing Null values; and
3. Regression models cannot deal with collinearity from columns that have the same data

````python
def missing_check(data):
    # Print out dataframe measuring missing values and their percentages
    
    missing = df.copy().isnull().sum()
    missing_pct = (missing / len(df) * 100).round(2)
    missing_df = pd.DataFrame({'count': missing, 'percent': missing_pct})[missing > 0]
    print("Missing values:\n", missing_df,"\n")

missing_check(df)
````

Nothing here gives us reason to believe that the data is Not Missing At Random

But recalling that the relative percent of the missing values is substantially low, subgrouping over the most reasonable conditions and imputing should have a low impact on the model as a whole.

These are both first class female passengers. We will then impute the missing values for 'embarked' and 'embark_town' with the most frequent class between Female passengers of First Class.

````python
# Compute the most common 'embarked' and 'embark_town' using the missing values conditions
embarked_fill = df.loc[((df['sex'] == 'female') & (df['pclass'] == 1)), 'embarked'].value_counts().idxmax()
print(f"Filling the missing entries in 'embarked' with {embarked_fill}")
embark_town_fill = df.loc[((df['sex'] == 'female') & (df['pclass'] == 1)), 'embark_town'].value_counts().idxmax()
print(f"Filling the missing entries in 'embark_town' with {embark_town_fill}")

# Fill the dataframe's columns with the appropriate values
values = {'embarked': embarked_fill, 'embark_town': embark_town_fill}
df = df.fillna(values)
missing_check(df)
````

````python
# Is 'embarked' the first letter of 'embark_town'?
(df['embark_town'].str[0] == df['embarked']).sum() == df.shape[0]
````

As for Age, which has a significantly higher proportion of missing values. We can rule out NMAR reasoning as we don't know how the data was truly generated, and collecting more data is not possible without significant real life effort.

To determine between Missing at Random and Missing Completely At Random, we perform permutation tests along all the valid columns and see if any of them have

````python
df.head()
````

````python
# Encode missingness into a Column
df['age_missing'] = df['age'].isna()
# Store Results
results = []

for col in df.columns:
    # Skip target columns
    if col in ['age', 'age_missing']:
        continue

    # Create Permutation Test DataFrame with only relevant columns
    temp = df[[col, 'age_missing']]

    # Computes the Observed Statistic:
    # Defined as the Variance in Missingness Rates
    # Acrosss the groups in 'col'
    obs_stat = (temp.groupby(col)['age_missing'].mean().var())

    # Perform Permutation Test
    # With the number of repeats N = 5000
    perm_stats = []

    for _ in range(5000):
        shuffled = np.random.permutation(temp['age_missing'])
        stat = (temp.assign(shuffled=shuffled).groupby(col)['shuffled'].mean().var())
        perm_stats.append(stat)

    # Calculate the P-Value of the Column
    # Save results in a format unpackable by pd.DataFrame
    p_value = np.mean(np.array(perm_stats) >= obs_stat)
    results.append({'column': col, 'observed_stat': obs_stat, 'p_value': p_value})

# Collect Results
results_df = (pd.DataFrame(results).sort_values('p_value').reset_index(drop=True))

print(results_df)
````

Our results tell us that age is definitely highly correlated with every column. 
Imputation using all statistically significant columns may seem effective here, but that runs the risk of creating subgroups with sizes too small to retrieve statistics from.
Instead, data from the most descriptive statistically significant columns will be used

````python
df['age'] = df.groupby(['who', 'pclass'])['age'].transform(lambda x: x.fillna(x.median()))
missing_check(df)
````

With no more

### Question 1: "Was the Women and Children First" Order Effective?"
First, let's analyze our target variable, 'survived'.

A main factor consistent with multiple other research reports was the use and effectiveness of the "Women and Children First" order given by Captain E.J. Smith. Given this in mind, let's see how much our dataset reflects this conclusion. 

````python
nonsurvivors, survivors = [x for _, x  in titanic.groupby(titanic['survived'] == 1)]

pivot_survival = titanic.pivot_table(index = 'sex', 
                                values = 'survived',
                                aggfunc = ['mean', 'sum', 'count']
)
# Note that the mean of the survival column returns the survival rate
# The sum returns the number of survivors
# and the count returns the number of passengers
pivot_survival.columns = ['survival rate', 'number of survivors', 'total passengers']
pivot_survival
````

This data shows a trend, but for ease of understanding, visualizing the data is best here.

````python
survivor_palette = ["#228b22","#b22222"]

# Plot the survival rate by itself
survivor_plot = sns.histplot(
    data = titanic,
    x = 'alive',
    hue = 'alive',
    hue_order = ['yes', 'no'],
    palette = survivor_palette,
    stat = 'probability',
    multiple = 'dodge',
    shrink = 0.8)
plt.xlabel("Alive?")
plt.ylabel("Proportion")
plt.ylim(0,1)
plt.title("Survival Data")

plt.show()
````

````python
# Plot the survival rate conditioning on gender
survivor_plot_gendered = sns.histplot(
    data=titanic, 
    x = "sex", 
    hue = 'alive', 
    common_norm = False, 
    multiple = 'dodge', 
    stat = 'probability', 
    hue_order = ['no', 'yes'], 
    palette = survivor_palette[::-1], 
    shrink = 0.8)
plt.title("Male vs Female Survival Data")
plt.ylabel("Proportion")
plt.ylim(0,1)
plt.show()
````

Comparing the graphs, as well as verifying with the numerical statistics from earlier, we can see that there was a significant correlation between female passengers and surviving the Titanic that goes against the non-conditioned statistics.

To verify the statistical significance of this conclusion, a Chi-Squared test is run below to retrieve its p value.

````python
chi_squared_sex = stats.chi2_contingency([
    [
        pivot_survival.loc['male', 'number of survivors'],
        pivot_survival.loc['male', 'total passengers']
        - pivot_survival.loc['male', 'number of survivors']
    ],
    [
        pivot_survival.loc['female', 'number of survivors'],
        pivot_survival.loc['female', 'total passengers']
        - pivot_survival.loc['female', 'number of survivors']
    ]
])
print("\nChi-Squared Test on 'sex' Categorical:")
print("Statistic: ", chi_squared_sex.statistic)
print("P-Value: ", chi_squared_sex.pvalue, '\n')
````

The significantly low p-value tells us to reject the null hypothesis of zero correlation. 

This process safelyh concludes that there is a statistically significant correlation between gender and survival rates. 

Moving on to Age, let's perform the same demographic tests, this time to confirm the "Children" portion of "Women and Children First"

````python
# Plot a binned graph of the Titanic Passenger Ages
age_plot = sns.histplot(
    data = titanic,
    x = 'age',
    multiple = 'dodge',
    legend = False
)
plt.ylim(0,100)
plt.title("Passenger Composition of the Titanic by Age")
plt.xlabel('Age')
plt.show()
````

Of some interest here is which way the graph skews. In other words, are lower or higher ages more likely to survive the Titanic. To test the dataset, a T-test must be performed

````python
ttest_age = stats.ttest_ind(survivors["age"], nonsurvivors["age"])
print("T-Test for Age:")
print("Statistic: ", ttest_age.statistic)
print("P-Value: ", ttest_age.pvalue)
````

The P-Value tells us that survivors were, on average, younger. 

Finally, let's add a new categorical to our dataset named 'is_child' to better capture the effect, since the rule was "Children first", not "Younger first". 



````python
titanic['is_child'] = titanic['age'] < 18

child_survivors = titanic[(titanic['is_child']) & (titanic['survived'] == 1)]
child_nonsurvivors = titanic[(titanic['is_child']) & (titanic['survived'] == 0)]
adult_survivors = titanic[(~titanic['is_child']) & (titanic['survived'] == 1)]
adult_nonsurvivors = titanic[(~titanic['is_child']) & (titanic['survived'] == 0)]

chi_squared_age = stats.chi2_contingency(
    [[child_survivors.shape[0], child_nonsurvivors.shape[0]],
     [adult_survivors.shape[0], adult_nonsurvivors.shape[0]]]
)

child_survival_rate = child_survivors.shape[0] / (
    child_survivors.shape[0] + child_nonsurvivors.shape[0]
)

adult_survival_rate = adult_survivors.shape[0] / (
    adult_survivors.shape[0] + adult_nonsurvivors.shape[0]
)
print("Child survival rate:", child_survival_rate)
print("Adult survival rate:", adult_survival_rate)
print("\nChi-Squared Test on 'child' Categorical:")
print("Statistic: ", chi_squared_age[0])
print("P-Value: ", chi_squared_age[1], '\n')
````

Given that children had a much higher survival rate, and that the Chi-Squared test performed on this returned a statistically significant P-value, This portion of the EDA can conclude that Children on the Titanic showed much higher survival rates than adults, showing the correct expected effect of the 'Women and Children First' evacuation policy.

Third class has the highest disparity between all three classses, with a nearly 3:1 ratio of nonsurvival to survival, respectively.

##### Is the Correlation Statistically Significant?

````python
ct_raw = pd.crosstab(titanic['pclass'], titanic['survived'])
chi2, p, dof, expected = stats.chi2_contingency(ct_raw)

n = ct_raw.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(ct_raw.shape) - 1)))

print(f"Chi2:      {chi2:.2f}")
print(f"P-value:   {p:.20f}")
````

Because both pclass and survived are categorical variables, we'll use the Chi-Square test of independence to look for significance. The result is extremely unambiguous, with Chi^2 = 102.89 and p = 0.000, providing strong evidence that passenger class and survival are not independent of 'embarked'. This EDA strongly supports the conclusion that passenger class was a significant and meaningful predictor of survival on the Titanic. However, class may not be an isolated factor, as ticket prices may have varied between where they were purchased.

### Feature Engineering

Taking into account all our Exploratory Data Analysis has performed, let's start to deploy models and test their accuracy scores, rates of accuracy, and rates of true positive-false positive using ROC-AUC.

If our models perform accurately, then we can conclude that the survival rate on the Titanic was a culmination of the features collected in the dataset.

First, let's prep our dataset further for Machine Learning

(This feature taken from github user ttsteiger, whose work is re-linked here https://ttsteiger.github.io/projects/titanic_report.html)

Let's engineer a feature that captures 'sibsp' (number of siblings and spouses) and 'parch' (number of parents and children).
A variable named 'family_size' will do. Let's define it as a datapoint's ```sibs + parch + 1 (include themselves)```

Let's also encode the embarking town. Recall that Southampton has the most entries in its category. This means it'll be an effective category to baseline the model on. 

Finally, let's encode 'is_male'.

````python
titanic_ml = titanic.copy()

titanic_ml['family_size'] = titanic_ml['sibsp'] + titanic_ml['parch'] + 1
titanic_ml['is_male'] = titanic_ml['sex'].map({'male': 1, 'female': 0})
titanic_ml['embarked_c'] = (titanic_ml['embarked'] == 'C').astype(int)
titanic_ml['embarked_q'] = (titanic_ml['embarked'] == 'Q').astype(int)
````

This section will be a large amount of code blocks, so let's define a useful report to summarize our model performance.

report_titanic will return the accuracy_score, ROC-AUC score, and the classification report of our models.

Note that we will be using an 80/20 test train split for the rest of this portion, something easily verifiable

````python
def report_titanic(features, model_type, y_train, y_pred_train, y_proba_train, y_test, y_pred_test, y_proba_test):
    print(f"Results on regression on 'titanic_ml' Dataset with {model_type}\n\n")
    print(f"Using features = {features}\n\n")
    
    print("Training dataset\n")
    print(f"Accuracy: {accuracy_score(y_train, y_pred_train):.3f}")
    print(f"ROC-AUC: {roc_auc_score(y_train, y_proba_train):.3f}")
    print(classification_report(y_train, y_pred_train))
    
    print("Testing dataset\n")
    print(f"Accuracy: {accuracy_score(y_test, y_pred_test):.3f}")
    print(f"ROC-AUC: {roc_auc_score(y_test, y_proba_test):.3f}")
    print(classification_report(y_test, y_pred_test))
    return
````

### Logistical Regression, Baseline

Logistical Regression, in this deployment, is a machine learning algorithm based on the statistical model of the same name. Here, it is used to classify survivors and nonsurvivors. This first implementation is a useful baseline, using all engineered nonduplicate predictors. It also does not use any hyperparameter optimization or feature selection. 

````python
# Select useful columns, create feature matrix and response vector
features_baseline = ['pclass', 'is_male', 'age', 'family_size', 'fare', 'embarked_c', 'embarked_q']
X_base = titanic_ml[features_baseline]
y_base = titanic_ml['survived']

# Split up train and test dataset for models
X_train_base, X_test_base, y_train_base, y_test_base = train_test_split(
    X_base, 
    y_base, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_base
)

scaler_base = StandardScaler()
X_train_base_scale = scaler_base.fit_transform(X_train_base)
X_test_base_scale = scaler_base.transform(X_test_base)

lr_base = LogisticRegression(max_iter=1000, random_state=42)
lr_base.fit(X_train_base_scale, y_train_base)

y_pred_train_base = lr_base.predict(X_train_base_scale)
y_pred_test_base = lr_base.predict(X_test_base_scale)
y_proba_train_base = lr_base.predict_proba(X_train_base_scale)[:,1]
y_proba_test_base = lr_base.predict_proba(X_test_base_scale)[:,1]

report_titanic(features_baseline,
               "Blind Logistic Regression", 
               y_train_base, 
               y_pred_train_base, 
               y_proba_train_base,
               y_test_base, 
               y_pred_test_base,
               y_proba_test_base)
````

With no further changes to our dataset features, we get a baseline Training Accuracy of 80.9% and Test Accuracy of 81%. 
Recall from earlier that we saw a skew in fare pricing, most obvious on our Cherbourg plot.

From this, let's test a log transform and see if this performs better than our base model.
