## Abstract

In this project, we investigated whether passenger characteristics on the Titanic affected survival and whether these factors could be used to predict survivability. We focused on variables such as gender, age, and passenger class, especially in the context of the “Women and Children First” policy. We also wanted to determine if survival outcomes were mostly explained by these characteristics or if there was still a large element of randomness.

To do this, we used and cleaned a dataset 891 passengers, removing missing values and preparing the data for analysis. We performed EDA and statistical tests, including chi-squared tests and t-tests, to examine relationships between variables and survival. We then built several machine learning models, including logistic regression and a support vector machine (SVM), using features like class, age, fare, and family size, and evaluated their performance using accuracy and ROC-AUC.

We found that gender and passenger class were the strongest predictors of survival. Female passengers and higher classes were significantly more likely to survive, supporting the impact of social norms and access to resources. Age had a smaller effect, with younger passengers slightly more likely to survive. Our models achieved moderate accuracy (around 78–83%), showing that survival was not random but also not perfectly predictable, suggesting that both structured factors and chance played a role.

## Background and Prior Work

The RMS Titanic needs little introduction. A massive Olympic class ocean liner stretching 882ft long and 175 feet high, its size is only overshadowed by its reputation of sinking on its maiden voyage. On April 15, 1912, near midnight, a starboard collision with an iceberg damaged and flooded too many internal compartments, and the Titanic started sinking. Despite having a capacity of around 3300 passengers, the amount of lifeboats on board were only enough for about 1100 passengers, and a "Women and Children First" protocol was established and generally carried out. Of the roughly 2200 on board the Titanic, only around 700 actually survived.

The dataset we are analyzing includes information on 891 passengers, including their passenger class, age, sex, and their accompaniment. Catastrophic events are usually considered indiscriminate, but research consistently shows that certain factors shape survival outcomes. These include categories and numericals such as socioeconomic status (SES), age, and gender.

### Similar studies:

A paper from Bowdoin College found that the Titanic, like many disasters, had "differential effects" in which survival and suffering were unequally distributed across class, age, and gender lines. Researchers found that first-class passengers had a survival rate of about 62%, compared to 41% in second class and only 25% in third class (Bowdoin), showcasing stark socioeconomic differences translated to better survival rates. They theorize that this was due to the higher status translating into better access to information, lifeboats, and life-saving opportunities. The paper also incorporates a diagram of the Titanic, showcasing that third class passengers were housed lower, second class in the lower middle, and first class in the upper middle sections of the boat. This diagram led them to believe that if the boat were sinking, the first-class passengers had more time to reach life-saving measures. The same image showcases that women and children had much higher rates of survival, with 75 and 50 percent, respectively, while men only had about 20% survival. This follows suit with the order Captain E.J. Smith gave to officers, "Women and Children First." The paper also mentions that certain officers interpreted these orders differently, with Officer Lightroller taking the order to mean women and children only, lowering lifeboats even before full capacity to prioritize women and children's survival.

In another analysis, Frey, Heggli, and Osterloh used individual-level Titanic data to test how economic resources, age, and gender affected the probability of survival. Aside from reconfirming that class and gender played a large role in the probability of survival, they showed age mattered as well. People in their "prime" (ages around 15-35) were more likely to survive than older passengers (Frey et al.), suggesting that physical mobility and stamina interacted with the other categories to shape outcomes.

A study similar to our proposed topic, but with a slightly different data set, concluded that gender was, in fact, the key determining factor in survivability. 

Because this is a historical topic, we can safely assume that there will be no issues with outdated data or analysis, but the data sets being analyzed may have varying results. Despite being the same size, with 891 passengers, the data set we are analyzing has some additional information that may reveal more nuance in the death statistics. Additionally, the individuals may be different, meaning a different population of the Titanic survivors was sampled.

An article from the NIH states that, despite the widespread belief of prioritizing women and children in maritime accidents, women in fact have a significant survival disadvantage in these situations It further asserts that despite the survival rate of women on the Titanic being significantly higher, in the grander scheme of maritime accidents, data actually implies the opposite.

### Question 1: "Was the Women and Children First" Order Effective?"
Note: The deck column is missing for 96% of passengers and won't be useful for this analysis. Additionally, decks were shared between the classes, so we will be omitting deck level from this analysis completely.

First, let's analyze our target variable, 'alive'.

A main factor consistent with multiple other research reports was the use and effectiveness of the "Women and Children First" order given by Captain E.J. Smith. Given this in mind, let's see how much our dataset reflects this conclusion. 

````python
nonsurvivors, survivors = [x for _, x  in titanic.groupby(titanic['survived'] == 1)]

female_survivors, male_survivors = [x for _, x  in survivors.groupby(titanic['sex'] == 'male')]
female_nonsurvivors, male_nonsurvivors = [x for _, x  in nonsurvivors.groupby(titanic['sex'] == 'male')]

print("")
print("Total Passengers (With Valid Age): ", titanic.shape[0], "\n")
print("Male Passengers: ", male_survivors.shape[0] + male_nonsurvivors.shape[0])
print("Female Passengers: ", female_survivors.shape[0] + female_nonsurvivors.shape[0], '\n')
print("Male Survivors: ", male_survivors.shape[0])
print("Female Survivors: ", female_survivors.shape[0], '\n')
print("Male Nonsurvivors: ", male_nonsurvivors.shape[0])
print("Female Nonsurvivors: ", female_nonsurvivors.shape[0], "\n")

print("Average Male Survivors Age: ", male_survivors['age'].mean().round(2))
print("Average Female Survivors Age: ", female_survivors['age'].mean().round(2))
print("Average Male Nonsurvivors Age: ", male_nonsurvivors['age'].mean().round(2))
print("Average Female Nonsurvivors Age: ", female_nonsurvivors['age'].mean().round(2), "\n")

print("Survivor Stats:\n", survivors[["age"]].describe().round(2), "\n")
print("Nonsurvivor Stats:\n", nonsurvivors[["age"]].describe().round(2),"\n")
````

Let's visualize this information first, to better inspect it.

````python
survivor_palette = ["#228b22","#b22222"]

survivor_plot = sns.histplot(
    data = titanic,
    x = 'alive',
    hue = 'alive',
    hue_order = ['yes', 'no'],
    palette = survivor_palette,
    stat = 'probability',
    multiple = 'dodge',
    shrink = 0.8)
plt.xlabel("Alive?")
plt.ylabel("Proportion")
plt.ylim(0,1)
plt.title("Survival Data")

plt.show()
````

````python
survivor_plot_gendered = sns.histplot(
    data=titanic, 
    x = "sex", 
    hue = 'alive', 
    common_norm = False, 
    multiple = 'dodge', 
    stat = 'probability', 
    hue_order = ['no', 'yes'], 
    palette = survivor_palette[::-1], 
    shrink = 0.8)
plt.title("Male vs Female Survival Data")
plt.ylabel("Proportion")
plt.ylim(0,1)
plt.show()
````

Comparing the graphs, as well as verifying with the numerical statistics from earlier, we can clearly see that there was a significant correlation between female passengers and surviving the titanic that goes against the aggregate pattern. We will verify this conclusion with a Chi-Squared test to retrieve its p-value

````python
chi_squared_sex = stats.chi2_contingency(
    [[male_survivors.shape[0], male_nonsurvivors.shape[0]], 
     [female_survivors.shape[0], female_nonsurvivors.shape[0]]]
)
print("\nChi-Squared Test on 'sex' Categorical:")
print("Statistic: ", chi_squared_sex.statistic)
print("P-Value: ", chi_squared_sex.pvalue, '\n')
````

The significantly lower p-value tells us to reject the null hypothesis of zero correlation. Further, the statistical effects i This is to say, there is a statistically significant correlation between gender and survival rates. 

Moving on to Age, let's perform the same demographic tests, this time to confirm the "Children" portion of "Women and Children First"

````python
age_plot = sns.histplot(
    data = titanic,
    x = 'age',
    multiple = 'dodge',
    legend = False
)
plt.ylim(0,100)
plt.title("Passenger Composition of the Titanic by Age")
plt.xlabel('Age')
plt.show()
````

````python
age_survivor_plot = sns.histplot(
    data = titanic[titanic['survived'] == 1],
    x = 'age',
    hue = 'survived',
    palette = [survivor_palette[0]],
    multiple = 'dodge',
    legend = False
)
plt.ylim(0,100)
plt.title("Titanic Survivor Compositon by Age")
plt.xlabel('Age')
plt.show()
````

Visual representation is unclear. It's hard to tell, so let's perform a t-test.

````python
ttest_age = stats.ttest_ind(survivors["age"], nonsurvivors["age"])
print("T-Test for Age:")
print("Statistic: ", ttest_age.statistic)
print("P-Value: ", ttest_age.pvalue)
````

The P-Value tells us that survivors were, on average, younger. Finally, let's add a new categorical to our dataset named 'is_child' to better capture the effect, since the rule was "Children first", not "Younger first".


````python
titanic['is_child'] = titanic['age'] < 18

child_survivors = titanic[(titanic['is_child']) & (titanic['survived'] == 1)]
child_nonsurvivors = titanic[(titanic['is_child']) & (titanic['survived'] == 0)]
adult_survivors = titanic[(~titanic['is_child']) & (titanic['survived'] == 1)]
adult_nonsurvivors = titanic[(~titanic['is_child']) & (titanic['survived'] == 0)]

chi_squared_age = stats.chi2_contingency(
    [[child_survivors.shape[0], child_nonsurvivors.shape[0]],
     [adult_survivors.shape[0], adult_nonsurvivors.shape[0]]]
)

print("\nChi-Squared Test on 'child' Categorical:")
print("Statistic: ", chi_squared_age[0])
print("P-Value: ", chi_squared_age[1], '\n')
````

Finally, we can safely conclude that both Women and Children did benefit from the employment of the "Women and Children First" order.

We can immediately see that First class has both a higher survival rate and around 40 more total survivors than Second and Third class each, while having close to the same number of passengers as Second class and roughly half as many passengers as Third class.

First Class passengers survived at a rate of 65.2%, which is roughly 2.5 times higher than Third Class passengers at 23.9%.

Also note that the difference from first to second class is roughly linear, as well as the difference from second to third class. This implies a linear relationship between the pclass and survival rates, which we can leverage for our model. This is because this now allows us to avoid needing to encode these variables, seeing as how the relationship is, again, linear.

##### Crosstab Heatmap of Survival Data

Next, we'll use a heatmap to see how the ratio of survival was affected in each class.

````python
ct = pd.crosstab(titanic['pclass'], titanic['survived'], normalize='index')
sns.heatmap(ct, annot=True, fmt='.2%', cmap='RdYlGn')
plt.title('Survival Rate by Class (% of each class)')
plt.show()
````

##### Is the Correlation Statistically Significant?

````python
ct_raw = pd.crosstab(titanic['pclass'], titanic['survived'])
chi2, p, dof, expected = stats.chi2_contingency(ct_raw)

n = ct_raw.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(ct_raw.shape) - 1)))

print(f"Chi2:      {chi2:.2f}")
print(f"P-value:   {p:.20f}")
````

Because both pclass and survived are categorical variables, we'll use the Chi-Square test of independence to look for significance. The result is extremely unambiguous, with Chi^2 = 92.90 and p = 0.000, proving that the probability of seeing these relationships between Class and Survival being essentially zero. This EDA strongly supports the conclusion that passenger class was a significant and meaningful predictor of survival on the Titanic. However, class may not be an isolated factor, as ticket prices may have varied between where they were purchased.

### Area of Concern - Possible Multicollinearity between Class, Fare and Embarked

Something that caught our attention while going over the dataset were three columns -> ```Class```, ```Fare``` and ```Embarked```

The economic state of the world in 1912 was vastly different than the one we have in 2026. One such difference is the information gap between different countries, cities or even ports. 

As such, we were curious how the makeup of 'class' was for each 'embarked' town, and if a combination of one or the two affected the 'fare' their ports payed.

Specifically, we want to know if different embarking locations were charged roughly the same. 

Let's define some null and alternates

Our null in this scenario is that embarking location didn't have an effect on class or fare

Our alternate is that some town or towns has a statistically significant difference in class or fare

This would mean some sort of collinearity exists with embarked, class and fare, with one or both perhaps being a proxy for one of the three variables, which could worsen our model performance.

Of the three embarking towns, of great note is the town Cherbourg, based in France.

### Feature Engineering

Taking into account all our Exploratory Data Analysis has performed, let's start to deploy models and test their accuracy scores, rates of accuracy, and rates of true positive-false positive using ROC-AUC

First, let's prep our dataset further for Machine Learning

Let's engineer a feature that captures 'sibsp' (number of siblings and spouses) and 'parch' (number of parents and children).
A variable named 'family_size' will do. Let's define it as a datapoint's ```sibs + parch + 1 (include themselves)```

Let's also encode the embarking town. Recall that Southampton has the most entries in its category. This means it'll be a useful baseline for our encoding. 

Finally, let's encode 'is_male', as we noticed earlier that again, a large amount of our dataset is male.

````python
titanic_ml = titanic.copy()

titanic_ml['family_size'] = titanic_ml['sibsp'] + titanic_ml['parch'] + 1
titanic_ml['is_male'] = titanic_ml['sex'].map({'male': 1, 'female': 0})
titanic_ml['embarked_c'] = (titanic_ml['embarked'] == 'C').astype(int)
titanic_ml['embarked_q'] = (titanic_ml['embarked'] == 'Q').astype(int)
````

This section will be a large amount of code blocks, so let's define a useful report to summarize our model performance.

report_titanic will return the accuracy_score, ROC-AUC score, and the classification report of our models.

Note that we will be using an 80/20 test train split for the rest of this portion, something easily verifiable

````python
def report_titanic(features, model_type, y_train, y_pred_train, y_proba_train, y_test, y_pred_test, y_proba_test):
    print(f"Results on regression on 'titanic_ml' Dataset with {model_type}\n\n")
    print(f"Using features = {features}\n\n")
    
    print("Training dataset\n")
    print(f"Accuracy: {accuracy_score(y_train, y_pred_train):.3f}")
    print(f"ROC-AUC: {roc_auc_score(y_train, y_proba_train):.3f}")
    print(classification_report(y_train, y_pred_train))
    
    print("Testing dataset\n")
    print(f"Accuracy: {accuracy_score(y_test, y_pred_test):.3f}")
    print(f"ROC-AUC: {roc_auc_score(y_test, y_proba_test):.3f}")
    print(classification_report(y_test, y_pred_test))
    return
````

### Logistical Regression, Blind

````python
features_baseline = ['pclass', 'is_male', 'age', 'family_size', 'fare', 'embarked_c', 'embarked_q']
X_base = titanic_ml[features_baseline]
y_base = titanic_ml['survived']

X_train_base, X_test_base, y_train_base, y_test_base = train_test_split(
    X_base, y_base, test_size=0.2, random_state=42, stratify=y_base
)

scaler_base = StandardScaler()
X_train_base_scale = scaler_base.fit_transform(X_train_base)
X_test_base_scale = scaler_base.transform(X_test_base)

lr_base = LogisticRegression(max_iter=1000, random_state=42)
lr_base.fit(X_train_base_scale, y_train_base)

y_pred_train_base = lr_base.predict(X_train_base_scale)
y_pred_test_base = lr_base.predict(X_test_base_scale)
y_proba_train_base = lr_base.predict_proba(X_train_base_scale)[:,1]
y_proba_test_base = lr_base.predict_proba(X_test_base_scale)[:,1]

report_titanic(features_baseline,
               "Blind Logistic Regression", 
               y_train_base, 
               y_pred_train_base, 
               y_proba_train_base,
               y_test_base, 
               y_pred_test_base,
               y_proba_test_base)
````

With no further changes to our dataset features, we get a pretty solid accuracy. The model also performs well on our test dataset.

Recall from earlier that we saw a skew in fare pricing, most obvious on our Cherbourg plot.

From this, let's test a log transform and see if this performs better than our base model.

### Logistical Regression with Log Transform

````python
titanic_ml['fare_log'] = np.log1p(titanic_ml['fare'])
features_log = ['pclass', 'is_male', 'age', 'family_size', 'fare_log', 'embarked_c', 'embarked_q']

X_log = titanic_ml[features_log]
y_log = titanic_ml['survived']

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.2, random_state=42, stratify=y_log)

scaler_log = StandardScaler()
X_train_log_scale = scaler_log.fit_transform(X_train_log)
X_test_log_scale  = scaler_log.transform(X_test_log)

lr_log = LogisticRegression(max_iter=1000, random_state=42)
lr_log.fit(X_train_log_scale, y_train_log)
y_pred_train_log  = lr_log.predict(X_train_log_scale)
y_pred_test_log = lr_log.predict(X_test_log_scale)
y_proba_train_log = lr_log.predict_proba(X_train_log_scale)[:,1]
y_proba_test_log = lr_log.predict_proba(X_test_log_scale)[:,1]

report_titanic(
    features_log, 
    "Logistical Regression w/Log Transform of Fare", 
    y_train_log, 
    y_pred_train_log, 
    y_proba_train_log,
    y_test_log, 
    y_pred_test_log,
    y_proba_test_log)
````

### Grid Search Logistical Regression, Hyperparametrized?

Let's try a different model to see if the performance is any better. We will also give C inputs for the model to test and see if it can hyperparametrize and perform well on the test dataset at the same time.

````python
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

titanic_grid = GridSearchCV(LogisticRegression(max_iter=1000, random_state = 42), param_grid, cv=5, scoring='roc_auc')
titanic_grid.fit(X_train_base_scale, y_train_base)
grid_best_est = titanic_grid.best_estimator_

print(f"Grid Search Best Parameters{titanic_grid.best_params_}", '\n')
print("Grid Search found these parameters for best regression", grid_best_est, '\n')

y_pred_train_grid = grid_best_est.predict(X_train_base_scale)
y_proba_train_grid = grid_best_est.predict_proba(X_train_base_scale)[:,1]
y_pred_test_grid = grid_best_est.predict(X_test_base_scale)
y_proba_test_grid = grid_best_est.predict_proba(X_test_base_scale)[:,1]
````

````python
report_titanic(
    "based on Hyperparameter Search",
    "Logistic Regression w/GridSearchCV's Best Model",
    y_train_base,
    y_pred_train_grid,
    y_proba_train_grid,
    y_test_base,
    y_pred_test_grid,
    y_proba_test_grid
)
````

````python
models_dict = {
    "Logistical Regression": (y_pred_train_base, y_pred_test_base),
    "LR With Fare Transform": (y_pred_train_log, y_pred_test_log),
    "GridSearchCV Logistical Regression": (y_pred_train_grid, y_pred_test_grid),
    "SVM": (y_pred_train_svm, y_pred_test_svm)
}

def plot_cm_ax(ax, y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1], ax=ax)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")
    ax.set_title(title)

fig, axes = plt.subplots(2, 2, figsize=(12,10))
axes = axes.flatten()
for ax, (model_name, (y_pred_train, _)) in zip(axes, models_dict.items()):
    plot_cm_ax(ax, y_train_base, y_pred_train, f"{model_name} - Train")
plt.tight_layout()
plt.show()

# Plotting test confusion matrices
fig, axes = plt.subplots(2, 2, figsize=(12,10))
axes = axes.flatten()
for ax, (model_name, (_, y_pred_test)) in zip(axes, models_dict.items()):
    plot_cm_ax(ax, y_test_base, y_pred_test, f"{model_name} - Test")
plt.tight_layout()
plt.show()
````
